```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.set_option("display.max_columns", 999)
pd.set_option("display.max_rows", 999)
import warnings
warnings.simplefilter(action = "ignore")
```

```python
df = pd.read_csv("final_df.csv")
df.drop(['Unnamed: 0'], axis =1, inplace = True)
```

```python
df.head(5)
```

```python
df["SalePrice"] = np.log(df['SalePrice'])
```

```python
Y = df[['SalePrice']]
df.drop(columns = ['SalePrice'], inplace = True)
X = df
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
```

```python
feature = SelectFromModel(Lasso(alpha = 0.005, random_state = 0))
```

```python
feat = feature.fit(X,Y)
```

```python
feat.get_support()
```

```python
data = df.columns[feat.get_support()]
```

```python
data
```

```python
X = df[data]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV, StratifiedKFold
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.30)
```

```python
from sklearn.ensemble import RandomForestRegressor
Regressor = RandomForestRegressor()
```

```python
params = {"n_estimators" : [int(x) for x in np.linspace(100,1000,10)],
         "criterion" : ["mae", "mse"],
         "max_features": ["auto", "log2", "sqrt"],
         "max_depth" : [int(x) for x in np.linspace(10,1000,10)],
         "min_samples_split": [1,2,3,4,5,6,7,8,9],
         "min_samples_leaf": [1,3,4,6,8]}
```

```python
rf_random = RandomizedSearchCV(estimator = Regressor, param_distributions = params, n_iter = 100, cv = 3, verbose = 2, n_jobs = -1, random_state = 0)
```

```python
rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
randomised = rf_random.best_estimator_
y_pred = randomised.predict(x_test)
rf_random.best_score_
```

It gives an accuracy of the 86.94%

```python
param_grid = {"n_estimators" : [rf_random.best_params_['n_estimators']-200,
                                rf_random.best_params_['n_estimators']-100,
                                rf_random.best_params_['n_estimators'],
                                rf_random.best_params_['n_estimators']+100,
                                rf_random.best_params_['n_estimators']+200],
         "criterion" : [rf_random.best_params_['criterion']],
         "max_features": [rf_random.best_params_['max_features']],
         "max_depth" : [rf_random.best_params_['max_depth']-50,
                       rf_random.best_params_['max_depth']-40,
                       rf_random.best_params_['max_depth']-30,
                       rf_random.best_params_['max_depth']-20,
                       rf_random.best_params_['max_depth']-10,
                       rf_random.best_params_['max_depth'],
                       rf_random.best_params_['max_depth']+10,
                       rf_random.best_params_['max_depth']+20,
                       rf_random.best_params_['max_depth']+30,
                       rf_random.best_params_['max_depth']+40],
         "min_samples_split": [rf_random.best_params_['min_samples_split']-1,
                              rf_random.best_params_['min_samples_split'],
                              rf_random.best_params_['min_samples_split']+1],
         "min_samples_leaf": [1,2,3]}
```

```python
grid_rf = GridSearchCV(estimator = Regressor, param_grid = param_grid, n_jobs = -1, cv = 5, verbose = 2)
```

```python
grid_rf.fit(x_train,y_train)
```

```python
grid_rf.best_params_
```

```python
grid_rf.best_score_
```

It gives an accuracy of 86.89%
